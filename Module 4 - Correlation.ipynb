# Correlation

```python
# Correlation - is relationship between 2 numeric variables, 
# usually use scatterplots to show correlation

# Where x - explanatory/independent variable,
# y - response/dependent variable
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_happiness.csv')
df.drop(columns = 'Unnamed: 0', inplace = True)
df
```

```python
# Correlation coefficient - measures linear correlation between two numeric sets of data,
# outputs number between -1 and 1

# Where 1 - positive correlation (x increases, y increases); 
# 0 - no correlation; 
# -1 - negative correlation (x increases, y decreses)
```

```python
# Scatterplot, to examine correlation between happiness_score and gdp_per_cap

sns.scatterplot(x = 'happiness_score', y = 'gdp_per_cap', data = df)
```

```python
# Scatterplot with a trendline

sns.lmplot(x = 'happiness_score', y = 'gdp_per_cap', data = df, ci = None)
```

```python
# Calculating Correlation coefficient; order doesn't matter

df['happiness_score'].corr(df['gdp_per_cap']) #0.7279733012222976
df['gdp_per_cap'].corr(df['happiness_score'])
```

```python
sns.scatterplot(x = 'gdp_per_cap', y ='social_support', data = df)

print(df['gdp_per_cap'].corr(df['social_support']))
# We have negative moderate correlation between gdp_per_cap and social_support
```

```python
sns.lmplot(x = 'life_exp', y = 'gdp_per_cap', data = df, ci = None)

print(df['life_exp'].corr(df['gdp_per_cap']))
# We have positive strong correlation between life_exp and gdp_per_cap
```
